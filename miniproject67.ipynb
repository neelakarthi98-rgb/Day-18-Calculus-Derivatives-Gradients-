## Finding the Minimum of a Function

```python
import numpy as np

```

//// Define the function///

```python
##formala:
def f(x):
    return x**4-3*x**3+2
def f_der(x):
    return 4*x**3-9*x**2
```

```python

# Hyperparameters
learning_rate = 0.01
epochs = 1000
```

```python
x = np.random.randn()
print(f"Starting x: {x:.3f}")
```

////using for loop///

```python
for epoch in range(epochs):
    grad = f_der(x)           # Compute derivative
    x = x - learning_rate * grad  # Update x
    # Stop early if derivative is very close to 0
    if abs(grad) < 1e-6:
        print(f"Gradient near zero, stopping at epoch {epoch}")
        break
    
    # Print progress every 100 steps
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: x = {x:.4f}, f(x) = {f(x):.4f}, grad = {grad:.4f}")

# 6. Final result
print(f"\nMinimum found at x = {x:.4f}, f(x) = {f(x):.4f}")
```
