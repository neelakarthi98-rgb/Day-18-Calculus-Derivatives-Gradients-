## Simple Gradient Descent for Linear Regression

////Implement a basic linear regression model from scratch using gradient descent.///

```python
import numpy as np
import matplotlib.pyplot as plt
```

/////Generate synthetic data:///

```python
np.random.seed(42)
x=np.linspace(0,10,50)
y=3*x+2+np.random.randn(50)*2  ##add some noise
```

////Define Loss Function: The Mean Squared Error (MSE)///

```python
#### formula for MSE=1/n ​i=1∑n​(yi​−yi​^​)2
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
```

////Define Gradients:///

```python
# 3. Initialize parameters (slope m and intercept b)
m = np.random.randn()  # random small value
b = np.random.randn()
```

////////Implement Gradient Descent Loop:///

```python
learning_rate = 0.01  ###Set a learning_rate and epochs.
epochs = 1000
```

```python
for epoch in range(epochs):
    # Predictions
    y_pred = m * x + b    
    
     ###Calculate the gradients of the MSE.   ###formula:

    dm = (-2 / len(x)) * np.sum(x * (y - y_pred))
    db = (-2 / len(x)) * np.sum(y - y_pred)

    ###update the parameter
    m = m - learning_rate * dm
    b = b - learning_rate * db
```

/////Print the loss every 100 epochs to see it decreasing.

```python
if epoch % 100 == 0:
        loss = mse(y, y_pred)
        print(f"Epoch {epoch}: m = {m:.3f}, b = {b:.3f}, Loss = {loss:.4f}")
```

////visualize///

```python
plt.figure(figsize=(8,6))
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, m * x + b, color='red', label='Best Fit Line')
plt.title("Linear Regression using Gradient Descent")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

# Final parameters
print(f"\nFinal slope (m): {m:.3f}")
print(f"Final intercept (b): {b:.3f}")
```
