## Visualizing a Derivative

////Plot a function and its derivative to get an intuitive feel for the concept.///

```python
import numpy as np
import matplotlib.pyplot as plt
```

//// Define a function///

```python
def f(x):                 ##formula f(x)=x^2
    return x**2
def f_der(x):              ##derivate of f'(x)=2x
    return 2*x
```

////create a array//

```python
x=np.linspace(-5,5,100)
```

////Calculate Outputs: Calculate the y values///

```python
y=f(x)
y_der=f_der(x)
```

////visualize///

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="f(x) = xÂ²", color="blue", linewidth=2)
plt.plot(x, y_der, label="f'(x) = 2x", color="red", linestyle="--", linewidth=2)

# Add axes lines, labels, and legend
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.title("Function and Its Derivative")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

# 6. Show the plot
plt.show()
```
