### Visualizing a 3D Gradient

///Plot a 3D surface for a function like f(x,y)=x2+y2 and a vector showing its gradient at a specific point.//

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def f(x,y):
    return x**2+y**2
def f_der(x,y):
    df_dx=2*x
    df_dy=2*y
    return df_dx,df_dy
```

```python
x=np.linspace(-2,2,10)
y=np.linspace(-3,3,10)
x,y=np.meshgrid(x,y)
z=f(x,y)

```

//showing its gradient at a specific point.//

```python
x0,y0=1,1
z0=f(x0,y0)
df_dx,df_dy=f_der(x0,y0)
```

///visualize**//

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x,y,z, cmap='viridis', alpha=0.8)

# Plot the gradient vector using quiver
ax.quiver(x0, y0, z0, df_dx, df_dy, 0, color='r', length=1, normalize=True, label='Gradient Vector')

# Highlight the point
ax.scatter(x0, y0, z0, color='red', s=50)

# Labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z = f(x, y)')
ax.set_title('3D Surface of f(x, y) = x + y with Gradient Vector')

plt.legend()
plt.show()
```
